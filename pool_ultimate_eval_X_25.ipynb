```python
# import numpy as np
# import matplotlib.pyplot as plt
# import pandas as pds
# import cv2
from ultralytics import YOLO
# import torch
# from IPython.display import Image
```

```python
pilot = "pilotPV_panels.v1i.yolov8-obb/data.yaml"
rzeszow = "rzeszowSolar panels seg.v2i.yolov8-obb/data.yaml"
synth = 'synth-pv.v1i.yolov8/data.yaml'
model_pt = "solarpanels_pools_yolov8l-p2_1024_v1.pt"
```

```python
# model = YOLO(model=model_pt, task="segment", verbose=True)  # explicitly
model = YOLO(model=model_pt, verbose=True)
model.eval()
```
