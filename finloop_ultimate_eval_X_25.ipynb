```python
# import numpy as np
# import matplotlib.pyplot as plt
# import pandas as pds
# import cv2
from ultralytics import YOLO
# import torch
# from IPython.display import Image
```

```python
rzeszow = "rzeszowSolar panels seg.v2i.yolov8/data.yaml"
model_pt = "best.pt"
```

```python
syth_to_finetune="synthetic_data/data.yaml"
labeled_to_finetune="labeled_fine_tunning.v4i.yolov8/data.yaml"
```

```python
model = YOLO(model=model_pt, task="segment", verbose=True)  # explicitly
model = YOLO(model=model_pt, verbose=True)
model.eval()
```

## pilot data
