```python
import numpy as np
import matplotlib.pyplot as plt
# import pandas as pds
# import cv2
from ultralytics import YOLO
# import torch
from IPython.display import Image
```

```python
pilot = "pilotPV_panels.v1i.yolov8-obb/pilotPV_panels.v1i.yolov8-obb/data.yaml"
pilot = "pilotPV_panels.v1i.yolov8-obb/data.yaml"
rzeszow = "rzeszowSolar panels seg.v2i.yolov8-obb/data.yaml"
synth = 'synth-pv.v1i.yolov8/data.yaml'
model_pt = "best.pt"
```

```python
# model = YOLO(model=model_pt, task="segment", verbose=True)
model = YOLO(model=model_pt, verbose=True)
model.eval()
```

## pilot data

```python
results = model.val(data=pilot, single_cls=True, batch=64, iou=0.7)
```

#### box

```python
results.box.mean_results() # Return mean of results, mp, mr, map50, map.
```

#### seg

```python
results.seg.mean_results() # Return mean of results, mp, mr, map50, map.
```

#### nbs

```python
results.summary(decimals=3)
```

```python
results.summary(decimals=2)
```
